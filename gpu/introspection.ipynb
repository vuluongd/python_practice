```python
import pyopencl as cl
```

pyopencl là wrapper của OpenCL cho Python → cho phép bạn lập trình GPU bằng Python.

```python
# print header
print('\n' + '=' *60 + '\nOpen CL Platforms and Devices\n')
# get platforms
for platform in cl.get_platforms():
    print('=' *60)
    print('Platform-name: '  + platform.name)
    print('Platform-vendor: '  + platform.vendor)
    print('Platform-version: ' + platform.version)
    print('Platform-profile: ' + platform.profile)

    #get devices
    for device in platform.get_devices():
        print('+''-' *56)
        print('Device-name: ' + device.name)
        print('Device-type: ' + cl.device_type.to_string(device.type))
        print('Device-vendor: ' + device.vendor)
        print('Device-version: ' + device.version)
        print('Device-profile: ' + device.profile)
        print('Device-max-compute-units: ' + str(device.max_compute_units))
        print('Device-max-clock-frequency [MHz]: ' + str(device.max_clock_frequency))
        print('Device-global-mem-size [MB]: ' + str(device.global_mem_size // (1024 * 1024)))
        print('Device-local-mem-size [kB]: ' + str(device.local_mem_size // 1024))
        print('Device-max-mem-alloc-size [MB]' + str(device.max_mem_alloc_size // (1024 * 1024)))
        print('Device-max-work-group-size : ' +str(device.max_work_group_size))
        print('Device-max-work-item-dimensions : ' + str(device.max_work_item_dimensions))
        print('Device-max-work-item-sizes : ' + str(device.max_work_item_sizes))
        print('Device-image-support : ' + str(device.image_support))
        print('Device-extensions : ' + str(device.extensions))
    print('=' *60 + '\n')
# A simple script to list OpenCL platforms and devices
# Author: Michael Hirsch,
# License: MIT
# Date: June 2024
```
